```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("creditcard_2023.csv")
data
```

```python
# Checking for missing values
data.isna().sum()
```

```python
# Checking for duplicate values
data.duplicated().sum()
```

```python
# Extracting the independent feature
x = data.drop(['id','Class'],axis=1)
y = data.Class
x
```

```python
# Standardizing the data
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
x_scaled_df = pd.DataFrame(x_scaled,columns=x.columns)
x_scaled_df
```

```python
# Splitting the data into training (70%) and testing (30%)
x_train,x_test,y_train,y_test = train_test_split(x_scaled_df, y, test_size = 0.3, random_state = 42 ,stratify = y)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Training the regression model
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Defining a function for Checking Model Accuracy, Classification Report and Confusion Matrix
def model_eval(actual, predicted):
  acc_score = accuracy_score(actual, predicted)
  conf_matrix = confusion_matrix(actual, predicted)
  clas_rep = classification_report(actual, predicted)
  print('Model Accuracy is: ', round(acc_score, 2))
  print(conf_matrix)
  print(clas_rep)
```

```python
preds_model_train = model.predict(x_train)
preds_model_test = model.predict(x_test)
```

```python
print("Evaluation with Training Data: ")
model_eval(y_train, preds_model_train)
```

```python
print("Evaluation with Testing Data: ")
model_eval(y_test, preds_model_test)
```
